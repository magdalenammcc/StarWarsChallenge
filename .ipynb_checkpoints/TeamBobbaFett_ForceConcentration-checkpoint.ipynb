```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('galaxies.csv')
df.head(10)
```

```python
model=KMeans(n_clusters=3)
model.fit(df)
```

```python
N = 1000
colors = np.linspace(0,1,N)
my_cmap = sns.light_palette("Navy", as_cmap=True)
df.plot(kind='scatter',x='X',y='Y',c=colors, cmap=my_cmap)
plt.show()
```

```python
#we know there are 3 clusters, so we create Kmean with 3 clusters
#and fit it to our data in our dataframe
Kmean = KMeans(n_clusters=3)
Kmean.fit(df)
```

```python
#these are the centroids, now we find the position of the uppermost (max Y)
aux_matrix=Kmean.cluster_centers_
aux_matrix=aux_matrix[aux_matrix[:,1].argsort()[::-1]]
galaxy_coordinates=aux_matrix[0]
galaxy_coordinates
print (f'Our galaxy coordinates are: {galaxy_coordinates}')
```

```python
df.plot(kind='scatter',x='X',y='Y',c=colors, cmap=my_cmap)
plt.plot(Kmean.cluster_centers_[0,0], Kmean.cluster_centers_[0,1], 'ro')
plt.plot(Kmean.cluster_centers_[1,0], Kmean.cluster_centers_[1,1], 'ro')
plt.plot(Kmean.cluster_centers_[2,0], Kmean.cluster_centers_[2,1], 'ro')
```

```python
#we add a column to the dataframe so we can know the cluster that 
#every point belongs to
df['kmean'] = Kmean.labels_
df.head()
```

```python
#Generate a map only of the uppermost galaxy
#we can plot our dataframe with kmean value 0 (our galaxy is cluster 0)
df_galaxy=df[df.kmean == 0]

df_galaxy.plot(kind='scatter',x='X',y='Y')
```

```python
#we think Baby Yoda is in the rightmost planet of this galaxy.
#Plot the highlighted planet on the original data
rightmost_value = df_galaxy['X'].max()
Y_coordinate_planetDf=(df_galaxy[df_galaxy['X'] == rightmost_value]) #this brings a 1 line dataframe
Y_coordinate_planetArray=Y_coordinate_planetDf.to_numpy() #turn the df to a numpy array
Y_coordinate_planet=Y_coordinate_planetArray[0,1] #select X coordinate from numpy array
X_coordinate_planet=rightmost_value
planet_coordinates=[X_coordinate_planet,Y_coordinate_planet]
#Y_coordinate_planetArray
planet_coordinates
```

```python
df.plot(kind='scatter',x='X',y='Y',c=colors, cmap=my_cmap)
plt.plot(X_coordinate_planet, Y_coordinate_planet, 'ro')
```
